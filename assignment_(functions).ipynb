```python

balance = 1000

def login():
    u = input("Username: ")
    p = input("Password: ")
    return u == "admin" and p == "1234"

def deposit():
    global balance
    amt = float(input("Enter deposit amount: "))
    balance += amt
    print("Balance:", balance)

def withdraw():
    global balance
    amt = float(input("Enter withdraw amount: "))
    if amt <= balance:
        balance -= amt
        print("Balance:", balance)
    else:
        print("Insufficient balance")

def check_balance():
    print("Current Balance:", balance)

def main():
    if login():
        print("Login Successful")
        while True:
            print("\n1.Deposit  2.Withdraw  3.Check Balance  4.Exit")
            choice = input("Enter choice: ")

            if choice == "1":
                deposit()
            elif choice == "2":
                withdraw()
            elif choice == "3":
                check_balance()
            elif choice == "4":
                print("Thank you!")
                break
            else:
                print("Invalid choice")
    else:
        print("Wrong Login Details")

main()
```
