```python
username = "admin"
password = "1234"
balance = 0

u = input("Enter username: ")
p = input("Enter password: ")

if u == username and p == password:
    print("Login Successful\n")

    while True:
        print("\n1.Deposit  2.Withdraw  3.Check Balance  4.Exit")
        choice = input("Choose option: ")

        if choice == "1":
            amt = float(input("Enter amount: "))
            balance += amt
            print("Amount Deposited")

        elif choice == "2":
            amt = float(input("Enter amount: "))
            if amt <= balance:
                balance -= amt
                print("Amount Withdrawn")
            else:
                print("Insufficient Balance")

        elif choice == "3":
            print("Current Balance:", balance)

        elif choice == "4":
            print("Thank you!")
            break

        else:
            print("Invalid Choice")

else:
    print("Wrong Username or Password")
```

Simple Banking System – Explanation

This program is a basic banking system written in Python.

First, it stores a fixed username, password, and sets the balance to 0.

The user is asked to enter a username and password.
If both are correct, login is successful.
If not, it shows "Wrong Username or Password".

After login, a menu is shown with 4 options:

1. Deposit – The user enters an amount, and it is added to the balance.

2. Withdraw – The user enters an amount, and it is subtracted from the balance (only if enough money is available).

3. Check Balance – It shows the current balance.

4. Exit – The program stops.

The program keeps running until the user chooses Exit.

2.Input and Output

4.While loop

```python
import math


class Shape:
    def __init__(self, length, width=0):
        self.length = length
        self.width = width

    def get_area(self):
        return 0

    def display(self):
        print(f"Shape: Length = {self.length}, Width = {self.width}")



class Rectangle(Shape):
    def __init__(self, length, width):
        super().__init__(length, width)

    def get_area(self):
        return self.length * self.width

    def display(self):
        print(f"Rectangle: Area = {self.get_area()}")



class Circle(Shape):
    def __init__(self, radius):
        super().__init__(radius, 0)
        self.radius = radius

    def get_area(self):
        return math.pi * self.radius * self.radius

    def display(self):
        print(f"Circle: Radius = {self.radius}, Area = {self.get_area()}")


def main():
    shapes = [Rectangle(5, 4), Circle(3)]

    for shape in shapes:
        shape.display()


if __name__ == "__main__":
    main()
```

```python
class Member:
    def __init__(self, name, membershipID):
        self.name = name
        self.membershipID = membershipID

    def getBorrowLimit(self):
        return 0

    def display(self):
        print(f"Member: Name = {self.name}, ID = {self.membershipID}")


class RegularMember(Member):
    def __init__(self, name, membershipID):
        super().__init__(name, membershipID)

    def getBorrowLimit(self):
        return 3

    def display(self):
        print(f"Regular Member: Name = {self.name}, "f"Borrow Limit = {self.getBorrowLimit()}")



class PremiumMember(Member):
    def __init__(self, name, membershipID):
        super().__init__(name, membershipID)

    def getBorrowLimit(self):
        return 10

    def display(self):
        print(f"Premium Member: Name = {self.name}, "f"Borrow Limit = {self.getBorrowLimit()}, "f"Benefit = 10% Discount")



def main():
    members = [RegularMember("jagadeesh", 101),
               PremiumMember("suriya", 202)]

    for member in members:
        member.display()


if __name__ == "__main__":
    main()
```
